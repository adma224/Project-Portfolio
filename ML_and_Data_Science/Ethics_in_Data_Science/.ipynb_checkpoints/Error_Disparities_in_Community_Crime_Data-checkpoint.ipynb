## Observing Error Disparities 

In this section, we will explore the disparities in error for different "groups" of the dataset. 

### **What is Error Disparity in Machine Learning?**

**Error disparity** refers to differences in prediction errors across different **subgroups** of a dataset. It is an important metric in **fairness analysis**, as it helps identify whether a model performs **consistently across different demographic groups**, or whether it exhibits **bias**.

If a model makes significantly **higher errors for one group than another**, it suggests **unequal performance**, which can lead to **unfair and discriminatory outcomes**.

The error disparity between two test sets with errors $\epsilon_1$ and $\epsilon_2$ is $|\epsilon_1 - \epsilon_2|$. 

where **Error(A)** and **Error(B)** can be any error metric, such as:
- **Overall prediction error** (e.g., misclassification rate)
- **False Positive Rate (FPR)**
- **False Negative Rate (FNR)**

### **Example in Crime Prediction**
Suppose we train a model to predict **violent crime rates** in different communities. If we calculate the **false negative rate (FNR)** separately for communities with:
- **High Black Population** → $FNR = 0.05$
- **Low Black Population** → $FNR = 0.12$

The **error disparity** between these two groups would be:

$$
| 0.05 - 0.12 | = 0.07
$$

This indicates that the model **underestimates violent crime rates more frequently** in communities with a **lower Black population**, meaning it may allocate **more police resources** incorrectly.




### **Expected Error Disparities in the Dataset**

#### **Attributes Expected to Have High Error Disparity**
1. **`racePctBlack` (Percentage of Black Population in the Community)**  
   - **Reason**: As seen in our fairness analysis, models often **exhibit bias** when predicting crime-related outcomes across different racial groups. Historical data might contain **structural biases** in policing or crime reporting, leading to **different prediction errors** for communities with different racial compositions.
   - **Expected Effect**: The model may **overestimate crime rates** in communities with a high Black population or **underestimate** crime in lower Black population areas.

2. **`PctPopUnderPov` (Percentage of Population Under Poverty Line)**  
   - **Reason**: Economic disparities influence both crime rates and law enforcement policies. If the model is trained on historical data that includes **socioeconomic biases**, it might **predict higher crime rates in poorer communities**, even if actual crime levels are similar across different income levels.
   - **Expected Effect**: The model might **over-predict violent crime** in lower-income areas while **under-predicting** it in wealthier areas, leading to **resource misallocation** in policing and policy decisions.

#### **Attributes Expected to Have Low Error Disparity**
1. **`medIncome` (Median Household Income in the Community)**  
   - **Reason**: While income affects crime rates, it is a **continuous numerical variable**, which means that **small variations** in this feature **may not significantly alter model predictions** in the same way categorical variables (e.g., race, policing policies) do.
   - **Expected Effect**: Since income is **relatively well-distributed** across different communities, the model's errors are likely to be **consistent** across different income levels.

2. **`PctHousOccup` (Percentage of Housing Units that are Occupied)**  
   - **Reason**: This is a **structural factor** that is **less directly tied to criminal behavior** than racial or economic factors. Housing occupancy may **correlate weakly with crime rates**, but it is unlikely to be a **major driver** of model prediction errors.
   - **Expected Effect**: The model likely **makes similar errors** across different levels of housing occupancy, leading to **low error disparity**.



### Determining Error Disparity
Below is a function which takes in X and Y arrays, a column number, and a threshold. The function returns arrays X0 and Y0 containing all rows where the value in the specified column falls strictly below the threshold, as well as arrays X1 and Y1 containing all rows where the the value in the specified column is above or equal to the threshold. 

```python
def split_on_feature(X_test, Y_test, column, thresh):
    
    #Combines X_test and Y_test to view complete rows
    XY_test = np.append(X_test, y_test.reshape(994,1), axis=1)
    
    X0_test = XY_test[XY_test[:,column] < thresh]
    X1_test = XY_test[XY_test[:,column] >= thresh]
    
    #Divides it again into X_test and Y_test
    Y0_test = X0_test[:,-1].copy()
    X0_test = X0_test[:,:-1].copy()
    Y1_test = X1_test[:,-1].copy()
    X1_test = X1_test[:,:-1].copy()
    
    return (X0_test, X1_test, Y0_test, Y1_test) 
```

### Problem 5.2: Calculating All Discrepancies (10 points)

Now, let's evaluate the error disparities for the model we previously trained. For each feature in the dataset, we use the previous function to split on that column when the threshold is set to 0.5. Then compute the error disparity for the feature by calculating the error of predictions made on both X0 and X1. 

```python
y_hat = logreg.predict(X_test)
err_preds = np.zeros([122,])
df = pd.DataFrame({'Column':dataframe.columns[:-1],
                   'Disparity':err_preds})
for i in range(0, X_test.shape[1]):
    dispar_tuple = split_on_feature(X_test, y_test, i, 0.5)
    
    y_hat_X0 = logreg.predict(dispar_tuple[0])
    y_hat_X1 = logreg.predict(dispar_tuple[1])
    
    if (y_hat_X0.size>100 and y_hat_X1.size>=100):
        df.at[i,'Disparity'] = ( abs( error(y_hat_X0,dispar_tuple[2]) - error(y_hat_X1,dispar_tuple[3])) )
        
    #Intentionally leaving the columns where either of the splits have fewer than 100 rows as NaN
    else: df.at[i,'Disparity'] = np.nan 
    
df = df.sort_values(by='Disparity',ascending=False)

df.set_index('Column').dropna()
```

```python
# Drop NaN values and select top 20 features with highest disparity
df_clean = df.dropna().sort_values(by='Disparity', ascending=False).head(20)

# Plot bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x=df_clean['Disparity'], y=df_clean['Column'], palette='viridis')

# Labels and title
plt.xlabel("Error Disparity")
plt.ylabel("Feature")
plt.title("Top 20 Features with Highest Error Disparity")
plt.show()
```

### Problem 5.3: Other Types of Discrepancies (10 points)

Instead of error disparities, let's compute two other types of errors that are of interest to us: False Negative Disparity and False Positive Disparity. 

For the feature racePctblack (percentage of population that is African-American), which is in column 2, compute the False Positive rate and False Negative rate using the provided functions. You should threshold the feature at 0.5 as earlier to create the two sets of samples.

```python
## INPUTS:
# y - true labels
# y_hat - predicted labels
def fp_error(y, y_hat):
    fp_errors = [np.maximum(y_hat[i] - y[i], 0) for i in range(len(y))]
    return np.mean(fp_errors)

## INPUTS:
# y - true labels
# y_hat - predicted labels
def fn_error(y, y_hat):
    fn_errors = [np.maximum(y[i] - y_hat[i], 0) for i in range(len(y))]
    return np.mean(fn_errors)
```

```python
dis_tuple = split_on_feature(X_test, y_test, 2, 0.5)

y_hat_X0 = logreg.predict(dispar_tuple[0])
y_hat_X1 = logreg.predict(dispar_tuple[1])

y0_fperr = fp_error(dispar_tuple[2],y_hat_X0)
y0_fnerr = fn_error(dispar_tuple[2],y_hat_X0)
y1_fperr = fp_error(dispar_tuple[3],y_hat_X1)
y1_fnerr = fn_error(dispar_tuple[3],y_hat_X1)

print('False Positive Error Rate of Communities with Above Median Black Population: ', y1_fperr)
print('False Positive Error Rate of Communities with Below Median Black Population: ', y0_fperr)

print('False Negative Error Rate of Communities with Above Median Black Population: ', y1_fnerr)
print('False Negative Error Rate of Communities with Below Median Black Population: ', y0_fnerr)
```

### **Explanation of  Results and Model Implications on Fairness**  

We examinied **False Positive and False Negative Disparities** across different groups within the dataset. Specifically, we analyzed the effect of the feature **racePctBlack** (percentage of the population that is African-American) on prediction errors.

#### **Key Observations from the Results**
1. **False Positive Rate Disparity**  
   - The **False Positive Rate (FPR)** for communities with a **higher percentage of Black residents** was **0.0**, while for those with a **lower percentage** it was **0.0425**.
   - This means the model **never incorrectly classified a non-violent community as violent** in areas with a high Black population, but it did make such errors in communities with a lower Black population.

2. **False Negative Rate Disparity**  
   - The **False Negative Rate (FNR)** for communities with **higher Black population** was **0.0**, while for those with a **lower percentage** it was **0.1093**.
   - This suggests that the model **never missed identifying violent communities** in high-Black population areas, but it frequently **underestimated** violence in lower-Black population areas.

#### **Implications on Fairness**
These disparities indicate **potential bias in the model**. Several fairness concerns arise:

1. **Unequal Treatment of Different Communities**  
   - The model appears to **over-predict violent crime** in communities with a lower Black population (higher False Negatives), while **never making such mistakes in higher Black population areas**.
   - This could indicate **imbalanced training data**, where the model learned to make different types of errors for different groups.

2. **Possible Discriminatory Outcomes**  
   - If the model is being used for policy decisions, such as **police resource allocation or crime prevention efforts**, it could result in **over-policing or under-policing** in certain areas.
   - For example, a model that **underestimates violence** in predominantly white communities could lead to **less law enforcement presence** than needed, while an over-predictive model in Black communities could contribute to **excessive policing**.

3. **Bias Amplification in Real-World Applications**  
   - If this model were used in **predictive policing**, it could reinforce existing biases by **misallocating police resources** based on flawed predictions.
   - Systematic bias could lead to **disproportionate targeting** of certain communities, even if actual crime rates are similar.

4. **Lack of Transparency in Decision-Making**  
   - If stakeholders are unaware of these disparities, they may **trust model predictions blindly**, leading to **unjust policies and resource allocations**.
   - Transparency is crucial to ensure that the model does not **unintentionally reinforce social biases**.

#### **How to Address These Issues**
To mitigate unfairness, the following actions could be taken:
- **Re-examine the training data** to check for imbalanced representations of different demographic groups.
- **Apply fairness-aware techniques**, such as **equalized odds** or **disparate impact analysis**, to ensure the model does not disproportionately affect certain populations.
- **Investigate alternative model architectures** that reduce bias, such as **fair regularization** or **adversarial debiasing**.
- **Engage domain experts** and policymakers to interpret model predictions responsibly and avoid reinforcing existing inequalities.
